```
from pyspark.sql.functions import regexp_replace, trim
df = df.withColumn("grau_de_instrucao", regexp_replace("grau_de_instrucao", '4ª série completo', 'ensino fundamental incompleto'))
#df = df.withColumn("grau_de_instrucao", regexp_replace("grau_de_instrucao", 'ensino fundamental completo', 'ensino fundamental completo'))
df = df.withColumn("grau_de_instrucao", regexp_replace("grau_de_instrucao", 'ensino médio comp', 'ensino médio completo'))
df = df.withColumn("grau_de_instrucao", regexp_replace("grau_de_instrucao", 'ensino medio incompleto', 'ensino médio incompleto'))
df = df.withColumn("grau_de_instrucao", regexp_replace("grau_de_instrucao", 'graduação', 'ensino superior completo'))

#df = df.withColumn("grau_de_instrucao", regexp_replace("grau_de_instrucao", 'completo', ''))

df = df.withColumn("grau_de_instrucao", trim(df.grau_de_instrucao))
df = df.withColumn("cargo", trim(df.cargo))
display(df)
```

```
import pyspark.sql.functions as F

df = df.withColumn('cargo', F.concat(F.upper(F.expr("substring(cargo, 1, 1)")), F.lower(F.expr("substring(cargo, 2)"))))

df = df.withColumn('grau_de_instrucao', F.concat(F.upper(F.expr("substring(grau_de_instrucao, 1, 1)")), F.lower(F.expr("substring(grau_de_instrucao, 2)"))))
df.show()
```

```
from pyspark.sql.types import DateType, TimestampType
from pyspark.sql.functions import *

df_tipos = df
df_tipos = df_tipos.withColumn(
    'mes_referencia',
    #F.date_format('mes_referencia', 'dd/MM/yyyy') #hh:mm:ss a
    #col('mes_referencia').cast(DateType())
    #to_date(df_tipos.mes_referencia, 'dd/MM/yyyy')
    #F.to_date(F.unix_timestamp(df_tipos.mes_referencia, 'dd/MM/yyyy').cast('timestamp'))
    #F.to_timestamp('mes_referencia', format='dd/MM/yyyy') #Ficou com minutos
    #unix_timestamp(df_tipos.mes_referencia, 'dd/MM/yyyy').cast(TimestampType()) #Ficou com minutos

    to_date(unix_timestamp(col('mes_referencia'), 'dd/MM/yyyy').cast("timestamp")) #Formato Favorito

    ###F.from_unixtime("mes_referencia",'yyyy/MM/dd').cast('timestamp') Não funcionou
    
    #to_date("mes_referencia", "dd/MM/yyyy") #yyyy/MM/dd
    #to_date("mes_referencia", "yyyy/MM/dd")
).withColumn(
    'id_funcionario',
    col('id_funcionario').cast('int')
).withColumn(
    'data_de_nascimento',
    #col('data_de_nascimento').cast('date')
    to_date(unix_timestamp(col('data_de_nascimento'), 'dd/MM/yyyy').cast("timestamp"))
).withColumn(
    'data_de_admissao',
    #col('data_de_admissao').cast('date')
    to_date(unix_timestamp(col('data_de_admissao'), 'dd/MM/yyyy').cast("timestamp"))
).withColumn(
    'data_de_demissao',
    #col('data_de_demissao').cast('date')
    to_date(unix_timestamp(col('data_de_demissao'), 'dd/MM/yyyy').cast("timestamp"))
).withColumn(
    'grau_de_instrucao',
    col('grau_de_instrucao').cast('string')
).withColumn(
    'cargo',
    col('cargo').cast('string')
).withColumn(
    'salario',
    col('salario').cast('double')
)

df_tipos.dtypes
```
